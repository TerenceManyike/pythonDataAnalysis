# Pivot tables for users and organizations

## Organization pivot table

### Getting the evolution of all employees number per country over ther years

```python
# Cleaning up null values from organizations data
org_csv_data.dropna(inplace = True)
```

```python
org_csv_data = org_csv_data.pivot_table(index='Founded', columns='Country',
                                        values='Number of employees', 
                                        aggfunc='sum')
# 
```

```python
org_csv_data
```

```python
org_csv_data = org_csv_data.iloc[:, :6] # Getting first six columns
org_csv_data
```

## Users pivot table

### Getting the total number of orders made by gender per job title

```python
# Cleaning up null values from users data
users_csv_data.dropna(inplace = True)
users_csv_data
```

```python
users_csv_data = users_csv_data.pivot_table(index='Sex', columns='Job Title',
                               values='orders',
                               aggfunc='sum')
users_csv_data # Displaying the pivot table
```
